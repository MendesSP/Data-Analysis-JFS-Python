##Import Libraries

```python
import graphlab as gl
```

##Load Data

```python
path = '/Users/andremendes/OneDrive/Estudar Projects/JFS/Python Data/'

train_data_complete = gl.SFrame.read_csv(path+'train_data_scores.csv')
train_data_results = gl.SFrame.read_csv(path+'train_data_results.csv')
train_data_num = gl.SFrame.read_csv(path+'train_data_num.csv')

test_data_complete = gl.SFrame.read_csv(path+'test_data_scores.csv')
test_data_results = gl.SFrame.read_csv(path+'test_data_results.csv')
test_data_num = gl.SFrame.read_csv(path+'test_data_num.csv')
```

##Build complete matrix with scores, numbers and results

```python
train_data_complete
```

```python
col_number = ['Age','Gender','Type School','School Year','GPA','Upload Transcript','Support for extra activities','English','know about Estudar','english test','english percentil']
col_name = ['AE1_p','AEDP1_p','AE2_p','AEDP2_p','AE3_p','AEDP3_p','HP 1_p','HP 2_p',
                  'HP 3_p','PAP_p','AE1_n','AEDP1_n','AE2_n','AEDP2_n','AE3_n','AEDP3_n',
                  'HP 1_n','HP 2_n','HP 3_n','PAP_n']
all_features = ['Age','Gender','Type School','School Year','GPA','Upload Transcript','Support for extra activities','English','know about Estudar','english test','english percentil','AE1_p','AEDP1_p','AE2_p','AEDP2_p','AE3_p','AEDP3_p','HP 1_p','HP 2_p',
                  'HP 3_p','PAP_p','AE1_n','AEDP1_n','AE2_n','AEDP2_n','AE3_n','AEDP3_n','HP 1_n','HP 2_n','HP 3_n','PAP_n']

train_data_complete['result']=train_data_results['0']
for num in col_number:
    train_data_complete[num]=train_data_num[num]
    
test_data_complete['result']=test_data_results['0']
for num in col_number:
    test_data_complete[num]=test_data_num[num]
    
```

##Normalize Data

```python
tr = train_data_complete.to_dataframe()
for col in all_features:
    tr[col] = tr[col].apply(lambda x: (x - tr[col].min()) / (tr[col].max() - tr[col].min()))
train_data_complete = gl.SFrame(tr)

tr = test_data_complete.to_dataframe()
for col in all_features:
    tr[col] = tr[col].apply(lambda x: (x - tr[col].min()) / (tr[col].max() - tr[col].min()))
test_data_complete = gl.SFrame(tr)
```

##Build Logistic Model Using training - Uper Bound

```python
lr_model_train = gl.logistic_classifier.create(train_data_complete, features=all_features,target='result')
lr_model_train.evaluate(train_data_complete, metric='roc_curve')
gl.canvas.set_target('ipynb')
lr_model_train.show(view='Evaluation')
```

##Build Logistic Model Using test - Real Result

```python
lr_model = gl.logistic_classifier.create(train_data_complete, target='result', features = all_features, validation_set = test_data_complete)
lr_model.evaluate(test_data_complete, metric='roc_curve')
gl.canvas.set_target('ipynb')
lr_model.show(view='Evaluation')
```
